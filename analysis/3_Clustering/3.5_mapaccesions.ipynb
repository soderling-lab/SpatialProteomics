```python
def ensure_dirs_exists(path):
    if "." in path:
	    path = os.path.dirname(path)
    if not os.path.exists(path):
        os.makedirs(path)
    return
```

```python
## drop sawme sequence duplicates 
res.drop_duplicates(subset=['Accession','Sequence'], inplace=True)
## if all phos for that accedsio no, just keep no
def check_phos(group):
    group_no = group[group['Phos'] == 'No']
    group_yes = group[group['Phos'] == 'Yes']
    if not group_no.empty:
        group_no = group_no.iloc[[0]]  # Keep the first 'No' entry as a representative
    return pd.concat([group_yes, group_no])

res = res.groupby('Accession', group_keys=False).apply(check_phos).reset_index(drop=True)
res
```

```python
#### phosphosite, - 7, +7 
def keep_phospho(annotatedseq):
    matches = re.findall(r"\[(.*?)\]", annotatedseq)
    if matches:
        for mat in matches:
            if "Phospho" not in mat:
                annotatedseq = annotatedseq.replace(mat, "").replace("[]", "")
            else:
                annotatedseq = annotatedseq.replace("[Phospho (STY)]", "*")
    return annotatedseq
def lower_site(seq):
    seq = seq.replace("_", "")
    star_index = seq.find('*') 
    if star_index != -1:
        seq = seq[:star_index -1] + seq[star_index-1].lower() + seq[star_index:]
    return seq
mres['Sequence'] = mres['Sequence'].apply(keep_phospho)
mres['Sequence'] = mres['Sequence'].apply(lower_site)

mres
```

```python
def extend_phosphosite_sequences(row, fasta_dict):
    def find_closest_index(mystringi, checkstringi):
        for i in mystringi:
            for c in checkstringi:
                if i < c:
                    ind1 = i
                    ind2 = c
                else:
                    None
        return ind1, ind2

    new_seqs = []
    if '*' in row['Sequence']:
        splits = row['Sequence'].split("*")
    else:
        return [row['Sequence']]  # Return in a list for consistency

    total_seq = fasta_dict.get(row['Accession'], None)
    if not total_seq:
        print("Could not find accession")
        return [row['Sequence']]  # Return original sequence if not found
    for i, split in enumerate(splits):
        if i < len(splits)-1:
            if len(splits[i+1]) < 4:
                postcheck = row['Sequence'][row['Sequence'].find(split) + len(split):].replace("*", "")
            else:
                postcheck = splits[i+1]
            matc = [match.start() for match in re.finditer(split.upper(), total_seq)]
            matc2 =  [match.start() for match in re.finditer(postcheck.upper(), total_seq)]
            s1, s2 = find_closest_index(matc, matc2)
            if len(split)>= 7:
                prefix = split[-7:] + '*'
            else:
                needed = 7 - len(split)
                # matc = [match.start() for match in re.finditer(split.upper(), total_seq)]
                # matc2 =  [match.start() for match in re.finditer(splits[i+1].upper(), total_seq)]
                prefix = total_seq[s1-needed:s1+(len(split))] + '*'
            if len(splits[i+1]) >=7:
                post = splits[i+1][:7]

            else:
                post = total_seq[s2: s2+7]
            new_seqs.append(prefix +post)
    new_seqs = list(filter(None, new_seqs))
    return new_seqs
sample = {"Accession": 'A2ALS5',
          "Sequence": 'SENSSTQs*S*PEMPTT'}


# extend_phosphosite_sequences(sample, fasta_dict)
# Apply the function to each row and get a Series of lists
lst = mres.apply(lambda row: extend_phosphosite_sequences(row, fasta_dict), axis=1)
# # Explode the lists into separate rows
exploded = lst.explode()
# # # Assign the exploded series to a new column in a copy of the original DataFrame
mrese = mres.loc[exploded.index].copy()
mrese['MySequence'] = exploded
mrese.reset_index(drop=True, inplace=True)
mrese
```
